```
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import math
```

Firstly, calculate the image gradient magnitudes and directions.

```
def findMagnitudeDirections(image):
  gx = np.gradient(image)[0]
  gy = np.gradient(image)[1]
  
  #find the magnitudes:
  gradient = np.zeros(gy.shape)
  for x in range(gx.shape[0]):
    for y in range(gx.shape[1]):
      temp_gradient = (gx[x][y] ** 2 + gy[x][y] ** 2) ** 0.5
      #thresholding small gradient magnitudes to 0 
      if temp_gradient < 2:
        gradient[x][y] = 0
      else:
        gradient[x][y] = temp_gradient
  
  #find the direction:
  direction = np.rad2deg(np.arctan(gy/gx))

  return gradient, direction
```

Secondly, reshape the image to fits to the grid.

```
def reshapeImage(image, cell_size):
  row = image.shape[0]
  col = image.shape[1]
  m = row % cell_size
  n = col % cell_size
  new_row = row - m
  new_col = row - n
  row_left = m // 2
  col_left = n // 2
  new_image = image[row_left:row_left+new_row, col_left:col_left+new_col]
  return new_image
```

Thirdly, form the orientation histogram.

```
def orientationHis(gradient, direction, cell_size):
  bins = [[-15, 15], [15, 45], [45, 75], [75, 105], [105, 135], [135, 165]]
  m = image.shape[0] // cell_size
  n = image.shape[1] // cell_size
  histogram = np.zeros((m, n, 6))
  for i in range(m):
    for j in range(n):
      for a in range(i*cell_size, (i+1)*cell_size):
        for b in range(j*cell_size, (j+1)*cell_size):
          for bin in range(6):
            if bins[bin][0] <= direction[a][b] < bins[bin][1]:
              histogram[i][j][bin] += gradient[a][b]
              break
  return histogram
```
